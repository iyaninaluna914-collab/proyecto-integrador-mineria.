Al mantener bajo control el impacto de los registros nulos imputados como "Unknown", aseguramos que las decisiones de inversión en licencias de contenido estén respaldadas por datos reales de preferencia.
3. **Reducción de Complejidad (PCA):** Mediante la reducción dimensional logramos identificar los ejes rectores del comportamiento del usuario (Engagement vs. Antigüedad), simplificando la estructura de datos sin perder información crítica del negocio.
